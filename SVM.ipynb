```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv("C:\\Users\\RAJESWARI\\Downloads\\iphone_purchase_records.csv")
print(dataset)
```

```python
print(dataset.isnull())
```

```python
X=dataset[['Age','Salary']]
Y=dataset['Purchase Iphone']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC; 
classifier=SVC(kernel='poly',random_state=0);
classifier.fit(X_train,Y_train);

y_pred=classifier.predict(X);
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix;
accuracy=accuracy_score(Y,y_pred);
conmatrix=confusion_matrix(Y,y_pred);
print("ACCURACY:",accuracy)
print("Confusion:",conmatrix)
```

```python
import matplotlib.pyplot as plt  
plt.figure(figsize=(8, 6))
plt.scatter(X_test[Y_test == 0]['Age'], X_test[Y_test == 0]['Salary'], color='red',marker='*', label='No iPhone Purchase')
plt.scatter(X_test[Y_test == 1]['Age'], X_test[Y_test == 1]['Salary'], color='Green',marker='.', label='iPhone Purchase')
plt.title("Age and Salary vs. iPhone Purchase (Test Data)")
# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
